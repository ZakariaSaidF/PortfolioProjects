```python
%%sql
SELECT *
  FROM facts
 LIMIT 5;
```

```python
%%sql
SELECT MIN(population), MAX(population), 
       MIN(population_growth), MAX(population_growth)
  FROM facts;
```

```python
%%sql
SELECT name, population
  FROM facts
 WHERE population in ((SELECT MIN(population) FROM facts),
                      (SELECT MAX(population) FROM facts)
                     );
```

```python
%%sql
SELECT MIN(population), MAX(population), 
       MIN(population_growth), MAX(population_growth)
  FROM facts
 WHERE name != 'World'; 
```

```python
%%sql
SELECT AVG(population), AVG(area)
  FROM facts;
```

```python
%%sql
SELECT name, population, area
  FROM facts
 WHERE population > (SELECT AVG(population) FROM facts)
       and 
       area < (SELECT AVG(area) FROM facts);
```

```python
# What country has the most people?
%%sql
SELECT name, MAX(population)
  FROM facts
 WHERE name != 'World'
```

```python
# What country has the highest growth rate?
%%sql
SELECT name, MAX(population_growth)
  FROM facts
 WHERE name != 'World'
```

```python
# Which countries have the highest ratios of water to land?
%%sql
SELECT name, MAX(CAST(area_water AS float) / CAST(area_land AS float))
  FROM facts
 WHERE name != 'World'
```

```python
# Which countries have more water than land?
%%sql
SELECT name, area_land, area_water
  FROM facts
 WHERE (area_land - area_water) <= 0;
```

```python
%%sql
SELECT name, birth_rate, death_rate, birth_rate - death_rate AS add_most
  FROM facts
 ORDER BY add_most DESC
 LIMIT 10;
```

```python
%%sql
SELECT name, birth_rate, death_rate, death_rate - birth_rate
  FROM facts
 ORDER BY death_rate - birth_rate DESC;
```

```python
%%sql
SELECT name, population, area, population / area
  FROM facts
 ORDER BY population / area DESC;
```
